```python
#Seprated independent and dependent features. Droppind last columns .taking it separately

x=df.iloc[:,:-1]   #  Independent features
y=df.iloc[:,-1]   # dependent features
```

```python
# From independent features, seprate categorical and numerical features.

x_nume= [ column for column in x.columns if x[column].dtypes!='O']
x_cate= [ column for column in x.columns if x[column].dtypes=='O']
```

```python
# Now Transformationa and Standardization on numerical features
x[x_nume]
```

```python
# Transforming data to remove skewness. We cannont use Box-Cox method as it no used on negative values.
x_transformed= power_transform( x[x_nume], method='yeo-johnson' )
```

```python
# Standardization  StandardScaling  0-1-->Apply fit and transform
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_scaled= sc.fit_transform(x_transformed)

x_scaled
```

```python
#Encoding of categorical feature "Region"

# Label Encoder

x[x_cate]
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
x_encoded=le.fit_transform(x[x_cate])
```

```python
x_encoded
```

```python
x=pd.concat([pd.DataFrame(x_encoded),pd.DataFrame(x_scaled)  ], axis=1)
```

```python
x
```

```python
y
```

```python
# The target variable is continuous values that is decimal values then it is the regression problem statement
```

```python
# For train test split, find best random_state 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
lr=LinearRegression()
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=i,test_size=.20)
    lr.fit(x_train,y_train)
    predict_train=lr.predict(x_train)
    predict_test=lr.predict(x_test)
    if round(mean_squared_error(y_train,predict_train)*100,1)==round(mean_squared_error(y_test,predict_test)*100,1):
        print("At random state ", i, "The model performance very well")
        print("At random state: ",i)
        print("Training Mean squared error is: ", round(mean_squared_error(y_train,predict_train)*100,2))
        print('Test Mean squared error is: ', round(mean_squared_error(y_test,predict_test)*100,2))
        print('X'*50,'\n')
        
        
```
